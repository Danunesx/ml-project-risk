## ENCONDER

```python
from sklearn.preprocessing import LabelEncoder

X = df.drop(['Project_ID', 'Risk_Management_Maturity', 'Tech_Environment_Stability', 'Change_Control_Maturity', 'Risk_Level'], axis = 1)
y = df['Risk_Level']

X
```

```python
nulos = X.isnull().sum()
nulos[nulos > 0]
```

```python
label = LabelEncoder()

for i in X.select_dtypes(exclude=['number']).columns:
    X[i] = label.fit_transform(X[i])

X
```

```python
base = X.copy()
base
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

## MODELO DE ML

```python
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay, confusion_matrix, classification_report
```

```python
from sklearn.model_selection import train_test_split

X_treino, X_pred, y_treino, y_pred = train_test_split(X, y, test_size=0.25, random_state=0)
```

## NAIVE BAYES

```python
from sklearn.naive_bayes import GaussianNB

nb = GaussianNB()
nb.fit(X_treino, y_treino)
```

```python
pred = nb.predict(X_pred)
pred
```

```python
accuracy_score(y_pred, pred)
```

```python
cm = confusion_matrix(y_pred, pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(colorbar=False, cmap='Blues');
```

## RANDOM FOREST

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
rf.fit(X_treino, y_treino)
```

```python
pred = rf.predict(X_pred)
pred
```

```python
y_pred
```

```python
accuracy_score(y_pred, pred)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression(random_state=1)
lr.fit(X_treino, y_treino)
```

```python
pred = lr.predict(X_pred)

accuracy_score(y_pred, pred)
```

## Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier

grad = GradientBoostingClassifier(n_estimators=300 ,criterion='squared_error', random_state=0, max_features='sqrt')
grad.fit(X_treino, y_treino)
```

```python
pred = grad.predict(X_pred)
```

```python
accuracy_score(y_pred, pred)
```

```python
X.shape, y.shape
```

```python
from sklearn.model_selection import cross_val_score, KFold

resultados = []

for i in range(30):
    kFold = KFold(n_splits=10, shuffle=True, random_state=i)

    model = GradientBoostingClassifier(n_estimators=300 ,criterion='squared_error', random_state=0, max_features='sqrt')
    scores = cross_val_score(model, X, y, cv=kFold)
    resultados.append(scores.mean())
```
